```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
Data_web = requests.get('https://www.payscale.com/research/US/School=Santa_Clara_University/Salary#by_Degree')
Data_web.content
html_data = BeautifulSoup(Data_web.content, 'html.parser')
html_data.prettify()
table = html_data.find('table')
table_body = table.find('tbody')
data1 = {
    'Majors' : [],
    'Salary' : [],
}
count =0
table_names = ['Majors', 'Salary' ]
for row in table_body.find_all('tr'):
    for cell in row.find_all('td'):   
        data1[table_names[count%2]].append( cell.get_text() )
        count = count+1;
    
import pandas as pd
Datascrap_1 = pd.DataFrame(data1)
Datascrap_1
```
