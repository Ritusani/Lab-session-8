## Using Beautifulsoup library to scrape data and Data cleaning:
Below is the code that I used to scrape the data using beautiful soup and clean my data. 

```python
import pandas as pd
```

```python
from bs4 import BeautifulSoup
import requests
```

```python
Data2 = requests.get('http://www.higheredublog.com/top-universities-for-ms-in-mis-in-usa/')
```

```python
html_data = BeautifulSoup(Data2.content, 'html.parser')
```

```python
print(html_data.prettify())
```

```python
html_data.title
```

```python
table = html_data.find('table')
print(table)
```

```python
table_body = table.find('tbody')
print(table_body)
```

```python
for row in table_body.find_all('tr'):
    for cell in row.find_all('td'):
        print(cell)
```

```python
data = {
    'Ranking' : [],
    'University' : [],
    'Program' : [],
}
count =0
table_names = ['Ranking', 'University' , 'Program' ]
for row in table_body.find_all('tr'):
    for cell in row.find_all('td'):   
        data[table_names[count%3]].append( cell.get_text() )
        count = count+1;
    
import pandas as pd
Datascrap = pd.DataFrame(data)
```

```python
Datascrap.to_csv('University_Ranking.csv', index=False, encoding='utf-8')
```
